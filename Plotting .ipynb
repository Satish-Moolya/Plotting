```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(a, b)
axs[0, 0].set_title('Axis [0, 0]')
axs[0, 1].plot(a, b, 'tab:orange')
axs[0, 1].set_title('Axis [0, 1]')
axs[1, 0].plot(a, b, 'tab:green')
axs[1, 0].set_title('Axis [1, 0]')
axs[1, 1].plot(a, b, 'tab:red')
axs[1, 1].set_title('Axis [1, 1]')
```

```python
fig,(a1,a2) = plt.subplots(1,2)
a1.plot(a)
a2.plot(b)
fig,(a3,a4) = plt.subplots(1,2)
a3.plot(a,b)
a4.plot(b,a)
plt.tight_layout()
plt.show()
```

```python
# First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# Create just a figure and only one subplot
fig,ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')
```

```python
fig,(ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y)
ax1.set_title('Simple plot')
ax2.scatter(x,y)
ax2.set_title('Scatter')
plt.suptitle('PLOT')
plt.tight_layout()
```

```python
fig,(ax1,ax2) = plt.subplots(2,1)
ax1.plot(x,y)
ax1.set_title('Simple plot')
ax2.scatter(x,y)
ax2.set_title('Scatter')
plt.suptitle('PLOT')
plt.tight_layout()
```
